```python
import os
import numpy as np
import pydicom
```

```python
# Specify the path to the main directory containing patient folders
main_directory = './small_train_images'
```

```python
# Dictionary to store image data
image_data = {}

# Iterate over patient folders
for study_id in os.listdir(main_directory):
    study_id_dir = os.listdir(f'./small_train_images/{study_id}')
    
    # Iterate over series folders for each patient
    for series_id in study_id_dir:
        series_id_dir = os.listdir(f'./small_train_images/{study_id}/{series_id}')
        # Initialize list to store image arrays for the series
        image_arrays = []
        
        # Iterate over DICOM files in the series folder
        for instance in series_id_dir:
            image_path = f'./small_train_images/{study_id}/{series_id}/{instance}'
            
            # Read DICOM file
            ds = pydicom.dcmread(image_path)
            
            # Extract pixel data as numpy array
            pixel_array = ds.pixel_array
            
            # Append pixel array to the list
            image_arrays.append(pixel_array)
        
        # Store image arrays in the dictionary with (patient_id, series_id) tuple as key
        image_data[(study_id, series_id)] = image_arrays
```

```python
# Example usage: Access image data for a specific (patient_id, series_id) tuple
example_patient_id = '4003253'
example_series_id = '702807833'
if (example_patient_id, example_series_id) in image_data:
    images = image_data[(example_patient_id, example_series_id)]
    for i, image in enumerate(images):
        print(f"Image {i+1} shape:", image.shape)
else:
    print("No images found for the specified (patient_id, series_id) tuple.")
```
